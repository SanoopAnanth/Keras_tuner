```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img,img_to_array
```

```python
img=load_img('cat.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
#this is not numpy array, it is python image library
type(img)
```

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=[0.5,1.5]
)
```

```python
img=img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3) # why 1 is because of batches
```

```python
i=0
# use flow to operate on single image which is in numpy array, use flow_from_directory to operate on multiple images
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='./aug'):
    i+=1
    if i==10:
        break
```
